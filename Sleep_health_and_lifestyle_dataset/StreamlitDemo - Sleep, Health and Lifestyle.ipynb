# Import the necessary libraries

```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Dataset

```python
path = kagglehub.dataset_download("uom190346a/sleep-health-and-lifestyle-dataset")

df = pd.read_csv(path + "/Sleep_health_and_lifestyle_dataset.csv")
```

# EDA

```python
df.info()
print(df.isnull().sum())
df.head()
```

```python
df.fillna("Unknown", inplace=True)
df.drop("Person ID", axis=1, inplace=True)

df.info()
print(df.isnull().sum())
df.head()
```

# Exploratory Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", context="talk")

plt.figure(figsize=(10, 6))
sns.histplot(df['Sleep Duration'], kde=True, color="skyblue")
plt.title('Distribution of Sleep Duration')
plt.xlabel('Sleep Duration')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Sleep Duration'], color="lightgreen")
plt.title('Boxplot of Sleep Duration')
plt.xlabel('Sleep Duration')
plt.show()

numeric_df = df.select_dtypes(include=['float64', 'int64'])
if not numeric_df.empty:
    plt.figure(figsize=(12, 8))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix of Numeric Variables')
    plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Quality of Sleep', hue='Quality of Sleep', palette="viridis", legend=False)
plt.title('Frequency of Quality of Sleep Ratings')
plt.xlabel('Quality of Sleep')
plt.ylabel('Count')
plt.show()
```
