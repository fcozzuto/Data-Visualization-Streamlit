## Streamlit Demo

To install Streamlit, we only need a single line in Python, using pip 

Then, we can start using Streamlit functionalities based on our needs, by importing it in our notebook and declaring it, like any other python module/library

```python
import kagglehub
import os
import pandas as pd
import numpy as np

path = kagglehub.dataset_download("ahmedmohamed2003/quality-of-life-for-each-country")
print(os.listdir(path))

df = pd.read_csv(path + "/Quality_of_Life.csv", header=0)
df.head()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
print(f"Head:\n{df.head()}\n")
print(f"Describe:\n{df.describe()}\n")
print(f"Sum of Missing Values:\n{numeric_df.isnull().sum()}\n")
print(f"Sum of 0 Values:\n{(numeric_df == 0).sum()}\n")
print(f"Sum of Negative Values:\n{(numeric_df < 0).sum()}")
```
